```python
import numpy as np
import pandas as pd
```

```python
pd.options.display.max_columns = 999
pd.options.display.max_rows = 999
```

```python
housing = pd.read_csv('./train.csv')
housing.head()
```

```python
#Loading the test data and naming it 'holdout'
holdout = pd.read_csv('./test.csv')
holdout.head()
```

```python
# This will help avoid discrepancies in the columns when we get_dummies
traintest = pd.concat([housing, holdout])
```

```python
traintest.tail(10)
# Pandas has automatically alphabetized the columns in the new dataframe. 
# We inspect the 'SalePrice' column and see that all the values are NaN,
# assuring us that the test set was placed at the bottom.
```

```python
traintest.shape
# Verifying that the new dataframe has the same number of columns as 'housing'
# and that the number of its rows is the sum of those of 'housing' (2051) and 'holdout' (879).
```

```python
# Inspecting the target column
traintest['SaleCondition'].value_counts()
```

```python
# Transforming the target column appropriately
traintest['SaleCondition'] = traintest['SaleCondition'].apply(lambda x: 1 if x=='Abnorml' else 0)
```

```python
# Getting the dummies for the rest of the columns
traintest_dums = pd.get_dummies(traintest, drop_first=True)
```

```python
# Dividing the dataframe back into the 'housing' and 'holdout' rows, now with the same dummies. 
housing_new = traintest_dums.iloc[:2051, :]
holdout_new = traintest_dums.iloc[2051:, :]
```

```python
# Checking for null values
housing_new.isnull().sum().sort_values(ascending=False)
```

```python
# Replacing each null value with the median of its column
housing_new = housing_new.apply(lambda x: x.fillna(x.median()))
```

```python
#These columns had too many null values to keep around. 
housing_new = housing_new.drop(['LotFrontage', 'GarageYrBlt', 'MasVnrArea'],axis=1)
```

```python
# Standardizing the predictors. 
# Leaving 'PID' out because it's an index and 'SalePrice' because it's not in the test data. 

from sklearn.preprocessing import StandardScaler

y = housing_new['SaleCondition']
X = housing_new.drop(['PID', 'SalePrice', 'SaleCondition'], axis=1)

ss = StandardScaler()
Xs = ss.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

# Creating a train-test split within the given 'train' data
Xs_train, Xs_test, y_train, y_test = train_test_split(Xs, y, test_size=.3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

#Setting up the parameter dictionary for a KNN gridsearch
knn_params = {
    'n_neighbors':[1,3,5,9,13,21,35, 55],
    'weights':['uniform','distance'],
    'metric':['euclidean','manhattan']
}

knn_gridsearch = GridSearchCV(KNeighborsClassifier(), knn_params, cv=5, n_jobs=4, verbose=1)

knn_gridsearch.fit(Xs_train, y_train)
```

```python
# Checking out the optimal hyperparameters

print knn_gridsearch.best_params_
print knn_gridsearch.best_score_
```

```python
# Checking the 'optimal' model against the baseline. 

best_knn = knn_gridsearch.best_estimator_
print best_knn.score(Xs_test, y_test)
print 'baseline:', 1 - np.mean(y_test)
```

```python
# The model's accuracy matched the baseline exactly. 
# Checking the confusion matrix, we confirm that the model simply classified everything as positive (i.e.,not 'Abnorml').

from sklearn.metrics import confusion_matrix

predictions = best_knn.predict(Xs_test)

confusion_matrix(y_test, predictions)
```

```python
# Perhaps a logistic regression would work better (read: at all).
from sklearn.linear_model import LogisticRegression

logreg_params = {
    'penalty':['l1', 'l2'],
    'solver' : ['liblinear'],
    'C': np.logspace (-5, 5, 100)
}

logreg_gridsearch = GridSearchCV(LogisticRegression(), logreg_params, cv=5, verbose=2, n_jobs=8)

logreg_gridsearch.fit(Xs_train, y_train)
```

```python
# Checking out the optimal hyperparameters

print logreg_gridsearch.best_params_
print logreg_gridsearch.best_score_
```

```python
best_logreg = logreg_gridsearch.best_estimator_
print best_logreg.score(Xs_test, y_test)
print 'baseline:', 1 - np.mean(y_test)
```

```python
predictions = best_logreg.predict(Xs_test)

confusion_matrix(y_test, predictions)
```
