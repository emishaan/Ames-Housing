```python
#Loading the test data and naming it 'holdout'
holdout = pd.read_csv('./test.csv')
holdout.head()
```

```python
holdout.shape
```

```python
# This will help avoid discrepancies in the columns when we get_dummies
traintest = pd.concat([housing, holdout])
```

```python
traintest.tail(10)
# Pandas has automatically alphabetized the columns in the new dataframe. 
# We inspect the 'SalePrice' column and see that all the values are NaN
# assuring us that the test set was placed at the bottom.
```

```python
traintest.shape
# Verifying that the new dataframe has the same number of columns as 'housing'
# and that the number of its rows is the sum of those of 'housing' (2051) and 'holdout' (879).
```

```python
# Getting the dummies
traintest_dums = pd.get_dummies(traintest, drop_first=True)
```

```python
# Dividing the dataframe back into the 'housing' and 'holdout' rows, now with the same dummies. 
housing_new = traintest_dums.iloc[:2051, :]
holdout_new = traintest_dums.iloc[2051:, :]
```

```python
housing_new.tail()['SalePrice']
```

```python
# This cell and the above are to verify that the split was made in the right spot. 
holdout_new.head()['SalePrice']
```

```python
# Checking for null values
housing_new.isnull().sum().sort_values(ascending=False)
```

```python
# Replacing each null value with the median of its column
housing_new = housing_new.apply(lambda x: x.fillna(x.median()))
```

```python
#These columns had too many null values to keep around. 
housing_new = housing_new.drop(['LotFrontage', 'GarageYrBlt', 'MasVnrArea'],axis=1)
```

```python
# Standardizing the predictors
from sklearn.preprocessing import StandardScaler
y = housing_new['SalePrice']
X = housing_new.drop('SalePrice', axis=1)

ss = StandardScaler()
Xs = ss.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

# Creating a train-test split within the given 'train' data
Xs_train, Xs_test, y_train, y_test = train_test_split(Xs, y, test_size=.3)
```

```python
from sklearn.linear_model import LinearRegression

# Instantiating a linear regression and fitting to the train
linreg = LinearRegression()
linreg.fit(Xs_train, y_train)
```

```python
from sklearn.model_selection import cross_val_score

# Cross-validating (5-fold) withing the training set 
# and printing the scores along with their mean and standard deviation
scores = cross_val_score(linreg, Xs_train, y_train, cv=5)
print scores
print scores.mean()
print scores.std()
```

```python
#  ^ These are HORRENDOUS scores. Time to try a ridge. 
```

```python
from sklearn.linear_model import Ridge, RidgeCV

#Instantiating and fitting a Ridge Regression
rr = RidgeCV()
rr.fit(Xs_train, y_train)
```

```python
# Searching for (and finding) the optimal penalty coefficient

ridge_alphas = np.logspace(0, 5, 20)
print ridge_alphas
ridge_alphas
optimal_ridge = RidgeCV(alphas=ridge_alphas, cv=10)
optimal_ridge.fit(Xs_train, y_train)

print optimal_ridge.alpha_
```

```python
# Cross-validating with the optimal alpha

ridge = Ridge(alpha=optimal_ridge.alpha_)

ridge_scores = cross_val_score(ridge, Xs, y, cv=10)

print ridge_scores
print np.mean(ridge_scores)
```

```python
# Fitting to the train and scoring on the test
ridge.fit(Xs_train, y_train)
ridge.score(Xs_test, y_test)
```

```python
from sklearn.linear_model import Lasso, LassoCV
# Now to try a lasso

optimal_lasso = LassoCV(n_alphas=500, cv=10, verbose=1)
optimal_lasso.fit(Xs_train, y_train)

print optimal_lasso.alpha_
```

```python
lasso = Lasso(alpha=optimal_lasso.alpha_)

lasso_scores = cross_val_score(lasso, Xs, y, cv=10)

print lasso_scores
print np.mean(lasso_scores)
```

```python
lasso.fit(Xs_train, y_train)
lasso.score(Xs_test, y_test)
```
